```julia
using Zygote, Plots
using Random
```

```julia
x = collect(1:0.25:100)
y = map(x->3x^2+5x+4, x)
Plots.plot(x, y, seriestype = :scatter)
```

```julia
f(x, w, w, b) = w*x^2 + w*x + b
```

```julia
loss(y, t) = (y - t)^2/2
```

```julia
 = 7e-8 # The steps length.
threshold = 1
Random.seed!(1234)
```

```julia
# w = 4; w = 6; b = 4;
w = rand(); w = rand(); b = rand();

println("(w, w, b) ~ ($w, $w, $b)")
println("----------------------------------")
grad_w_av = 0
grad_w_av = 0
grad_b_av = 0
while true
    net_loss = 0
    grad_w = 0
    grad_w = 0
    grad_b = 0
    for i in 1:1:length(x)
        t = f(x[i], w, w, b)
        net_loss += loss(t, y[i]) # NetLoss per epoch
        grad_w += - x[i]^2 * (y[i]-t)
        grad_w += - x[i] * (y[i]-t)
        grad_b += - (y[i]-t)
#         println("$i is the counter")
"""
    Updating params per data-item.
#         w = w - *grad_w
#         w = w - *grad_w
#         b = b - *grad_b
"""
    end
    grad_w_av = grad_w/length(x)
    grad_w_av = grad_w/length(x)
    grad_b_av = grad_b / length(x)
    net_loss = net_loss/length(x)
    w = w - *grad_w_av
    w = w - *grad_w_av
    b = b - *grad_b_av
    println("Loss: ", net_loss, "\n")
    
    if net_loss < threshold # Found the minima cost
        break
    end
    if net_loss > 1e10 # Overshooting
        println("Overshoot. Try again")
        break
    end
end
w, w, b
```

```julia
function forward(x, y, w, w, b)
    y = f(x, w, w, b)
    l = loss(y, y)
    return l
end
```

```julia
 = 9.8e-8
threshold = 1e15
```

```julia
w = 10*rand(); w = 10*rand(); b = 10*rand();
println("(w, w, b) = ($w, $w, $b)");
println("------------------------------------")
net_loss_r = 0
while true
    net_loss = 0
    grad_w = 0
    grad_w = 0
    grad_b = 0
    for i = 1:length(x)
        net_loss += forward(x[i], y[i], w, w, b)
        _,_,g_w,g_w,g_b = Zygote.gradient((a,b,c,d,e)->forward(a,b,c,d,e),x[i],y[i],w,w,b)
        grad_w += g_w
        grad_w += g_w
        grad_b += g_b
    end
    grad_w_av = grad_w/length(x)
    grad_w_av = grad_w/length(x)
    grad_b_av = grad_b/length(x)
    w = w - *grad_w_av
    w = w - *grad_w_av
    b = b - *grad_b_av

    println("Loss: ", net_loss)
    if net_loss < 0.5
        break
    end
    if net_loss > threshold
        break
    end
    net_loss_r = net_loss
end
println("(w, w, b) = ($w, $w, $b)");
```
